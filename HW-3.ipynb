```python
import time
import random

name = input("What is your name? ")
time.sleep(1) # wait 1 second
print("Welcome " + name)
time.sleep(1) # wait 1 second
word = "hangman"
guesses = ''

turns = 10 # define ten turns for guessing

while turns > 0:
    failed = 0
    for char in word: # for loop for searching character in selected word
        if char in guesses:
            print(char)
        else:
            print("_")
            failed += 1
    
    if failed == 0: # if every character is predicted true print player won
        print("You won")
        break
    print('')
    guess = input("guess a character:") #take guess input from user
    guesses += guess
    if guess not in word: #if guessed character doesn't exist on word -1 turns
        turns -= 1
        print("Wrong")
    print("You have", + turns, 'more guesses')
    if turns == 0: # if there is no turn left print you lose
        print("You Lose")
```
